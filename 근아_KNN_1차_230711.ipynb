```python
# 이미지 배열을 저장할 빈 리스트를 생성합니다.
image_arrays_w = []

for file_name in file_list_w:
    # .ipynb_checkpoints와 같은 불필요한 파일을 건너뛰기 위한 조건문
    if file_name.endswith(('.jpg', '.jpeg')):
        # 파일 경로를 생성합니다.
        img_path = os.path.join(dir_path_w, file_name)
        # 이미지를 열고 numpy 배열로 변환합니다.
        img = Image.open(img_path)
        img_array = np.array(img)
        # 배열을 리스트에 추가합니다.
        image_arrays_w.append(img_array)

# 결과를 확인합니다.
for img_arr in image_arrays_w:
    print(img_arr)
```

```python
# 이미지 배열을 저장할 빈 리스트를 생성합니다.
image_arrays_c = []

for file_name in file_list_c:
    # .ipynb_checkpoints와 같은 불필요한 파일을 건너뛰기 위한 조건문
    if file_name.endswith(('.jpg', '.jpeg')):
        # 파일 경로를 생성합니다.
        img_path = os.path.join(dir_path_c, file_name)
        # 이미지를 열고 numpy 배열로 변환합니다.
        img = Image.open(img_path)
        img_array = np.array(img)
        # 배열을 리스트에 추가합니다.
        image_arrays_c.append(img_array)

# 결과를 확인합니다.
for img_arr in image_arrays_c:
    print(img_arr)
```

### KNN은 input data를 2차원 배열 형태로 넣어야 한다고 합니다.
KNeighborsClassifier와 같은 scikit-learn의 기계 학습 모델들은 2차원 배열 형태의 입력 데이터를 기대합니다.      
현재 이미지 데이터는 3차원 (높이, 너비, 컬러 채널) 또는 그 이상의 차원을 가지고 있을 것입니다. 따라서 이를 2차원으로 재구성해야 합니다.     
이미지 데이터를 2차원으로 변환하는 가장 일반적인 방법은 각 이미지를 단일 행으로 "평탄화"하는 것입니다.      
각 픽셀의 색상 채널 값을 연속적인 열로 나열하여 이미지를 단일 행의 2D 배열로 변환합니다.       
이렇게 하면 각 행은 하나의 이미지를 나타내며, 행의 각 열은 이미지의 각 픽셀을 나타냅니다.

```python
# 이미지 데이터를 2차원으로 재구성
n_images, height, width, channels = image_arrays_w_c_np.shape
image_arrays_w_c_2d = image_arrays_w_c_np.reshape(n_images, height * width * channels)
```

```python
image_arrays_w_c_2d.shape
```

### train, test data로 나누기 위해 임의의 index 생성

```python
np.random.seed(42)
index = np.arange(100)
np.random.shuffle(index)
```

```python
print(index)
```

```python
### train_data는 70개, test_data는 30개(총 100개)
```

```python
train_input = image_arrays_w_c_2d[index[:70]]
train_target = labels[index[:70]]
```

```python
print(image_arrays_w_c_2d[83], train_input[0])
```

```python
# 원본데이터와 비교시 동일한 것 확인
```

```python
test_input = image_arrays_w_c_2d[index[70:]]
test_target = labels[index[70:]]
```

```python
### KNN 모델 학습
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
```

```python
kn.fit(train_input, train_target)
```

```python
kn.score(test_input, test_target)
```

```python
kn.predict(test_input)
```

```python
test_target
```

### knn 모델의 매개변수 n_neighbors를 조정해 봄

```python
kn10 = KNeighborsClassifier(n_neighbors=10)
kn10.fit(train_input, train_target)
kn10.score(test_input, test_target)
```

```python
kn49 = KNeighborsClassifier(n_neighbors=49)
kn49.fit(train_input, train_target)
kn49.score(test_input, test_target)
```
