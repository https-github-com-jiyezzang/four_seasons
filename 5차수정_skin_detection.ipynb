## (2) Skin Detection

- 피부 영역에 대한 기준: https://www.sciencedirect.com/science/article/abs/pii/S0262885620300573?via%3Dihub
(Zero-sum game theory model for segmenting skin regions, Author: Djamila Dahmani, Mehdi Cheref, Slimane Larabi)
위 논문에 따라 ANN을 이용한 피부 영역 검출을 시행한 결과 기준으로 삼음 (여러 인종의 피부색을 고려)  
```
  0<=H<=17 and 15<=S<=170 and 0<=V<=255

			and

    0<=Y<=255 and 135<=Cr<=180 and 85<=Cb<=135
```

````python
def Skin_Detection(img_path, name = "태연"):

  #Open a simple image
  img=cv2.imread(img_path)

  #converting from gbr to hsv color space
  img_HSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  #skin color range for hsv color space
  HSV_mask = cv2.inRange(img_HSV, (0, 15, 0), (17,170,255))
  HSV_mask = cv2.morphologyEx(HSV_mask, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))

  #converting from gbr to YCbCr color space
  img_YCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
  #skin color range for hsv color space
  YCrCb_mask = cv2.inRange(img_YCrCb, (0, 135, 85), (255,180,135))
  YCrCb_mask = cv2.morphologyEx(YCrCb_mask, cv2.MORPH_OPEN, np.ones((3,3), np.uint8))

  #merge skin detection (YCbCr and hsv)
  global_mask=cv2.bitwise_and(YCrCb_mask,HSV_mask)
  global_mask=cv2.medianBlur(global_mask,3)
  global_mask = cv2.morphologyEx(global_mask, cv2.MORPH_OPEN, np.ones((4,4), np.uint8))


  HSV_result = cv2.bitwise_not(HSV_mask) # 피부영역을 어둡게 표현하기 위해 bitwise_not연산 사용
  YCrCb_result = cv2.bitwise_not(YCrCb_mask)
  global_result=cv2.bitwise_not(global_mask)

  # Extract the skin regions from the original image
  HSV_skin_detected = cv2.bitwise_and(img, img, mask=HSV_mask)
  YCrCb_skin_detected = cv2.bitwise_and(img, img, mask=YCrCb_mask)
  global_skin_detected = cv2.bitwise_and(img, img, mask=global_mask)


  #show results
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_1_HSV_result.jpg",HSV_result)
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_2_YCrCb_result.jpg",YCrCb_result)
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_3_global_result.jpg",global_result)
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_4_img_HSV.jpg",img_HSV)
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_5_img_YCrCb.jpg",img_YCrCb)
  # 6, 7, 8이 핵심 이미지들
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_6_HSV_skin_detected.jpg",HSV_skin_detected)
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_7_YCrCb_skin_detected.jpg",YCrCb_skin_detected)
  cv2.imwrite("/content/drive/MyDrive/four_seasons/변환된사진모음/" + name + "_8_global_skin_detected.jpg",global_skin_detected)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

  return img_HSV, img_YCrCb, HSV_mask, YCrCb_mask, global_mask
````

````python
# skin 좌표 추출, YCrCb 혹은 HSV 픽셀값 추출
def coordinate_extraction(mask, img):
  skin = np.argwhere(mask == 255) # mask의 픽셀값이 255인 skin부분의 좌표 추출

  pixel_values = [] # YCbCr 혹은 HSV 픽셀값 저장을 위한 리스트

  for index in skin:
    y, x = index  #일반적으로 생각하는 가로 x, 세로 y로 표현하기 위해 위치 바꿈
    #print(f"x: {x}, y: {y}") # skin 범위 좌표값 출력(스압주의)
    pixel_value = img[y, x]
    pixel_values.append(pixel_value)

  for coord, pixel_value in zip(skin, pixel_values): # skin부분의 좌표값과 픽셀값 출력을 위한 함수
    x, y = coord
    #print(f"Coordinate: ({x}, {y}), Pixel Value: {pixel_value}")

  return pixel_values
````

````python
# 픽셀의 평균값 구하기
def pixel_means(pixel_values):
  # 열 별로 원소들을 더하기
  column_sums = np.sum(pixel_values, axis=0)

  # pixel_values를 NumPy 배열로 변환
  pixel_values = np.array(pixel_values)

  # 열 별로 평균 계산하기
  column_means = column_sums / pixel_values.shape[0]

  return column_means
````

````python
def get_dict(dir_path):
  file_list = os.listdir(dir_path)

  img_list = []
  result_list = []

  for file in file_list:
    img_path = os.path.join(dir_path, file)
    if os.path.isdir(img_path):  # 만약 디렉토리인 경우, 무시하고 다음 파일로 넘어감
      continue
    file_name = file
    file_name = os.path.splitext(file_name)[0] # 파일 이름과 확장자 분리
    img_HSV, img_YCrCb, HSV_mask, YCrCb_mask, global_mask = Skin_Detection(img_path, file_name)
    img_list.append([img_HSV, img_YCrCb, HSV_mask, YCrCb_mask, global_mask])

    # HSV
    HSV_pixel_values = coordinate_extraction(HSV_mask, img_HSV)
    HSV_pixel_means = pixel_means(HSV_pixel_values)
    #print(HSV_pixel_means)

    # YCrCb
    YCrCb_pixel_values = coordinate_extraction(YCrCb_mask, img_YCrCb)
    YCrCb_pixel_means = pixel_means(YCrCb_pixel_values)
    #print(YCrCb_pixel_means)

    # HSV + YCrCb
    global_HSV_pixel_values = coordinate_extraction(global_mask, img_HSV)
    global_HSV_pixel_means = pixel_means(global_HSV_pixel_values)
    #print(global_HSV_pixel_means)

    global_YCrCb_pixel_values = coordinate_extraction(global_mask, img_YCrCb)
    global_YCrCb_pixel_means = pixel_means(global_YCrCb_pixel_values)
    #print(global_YCrCb_pixel_means)

    result_list.append([HSV_pixel_means, YCrCb_pixel_means, global_HSV_pixel_means, global_YCrCb_pixel_means])

  return result_list
````
